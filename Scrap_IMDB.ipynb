```python
#Scrapping Movie data from IMDB (given the movie ID's)
import re
import urllib.request as ul 
from bs4 import BeautifulSoup
import pandas as pd

#Import file that containts the movie number(ID) in column 1
#IMDB has a Movie ID for each movie which needs to be put in the url

data = pd.read_csv('Movie_number.txt',delimiter=" #",header=None)  

Numbers = data[0]
for i in range(len(Numbers)):
    Numbers[i]=Numbers[i].strip()

#Metascore =list()
Revenue =list()
Budget=list()
Summary = list()
Director = list()
Writers = list()
Stars = list()
Production = list()

# loop for all movies
for q in Numbers:
    if q == 'NA':
        Revenue.append('NA')
        Budget.append('NA')
        Summary.append('NA')
        Director.append('NA')
        Writers.append('NA')
        Stars.append('NA')
        Production.append('NA')
        
    else:
        url = "http://www.imdb.com/title/tt"+q+"/"
        url_response=ul.urlopen(url,timeout=10)
        soup = BeautifulSoup(url_response,'html.parser')

        B_flag=0
        R_flag=0
        for h in soup.find_all('h4'):
            if "Budget:" in h:
                Budget.append(h.next_sibling.strip())
                B_flag=1
            if "Gross:" in h:
                Revenue.append(h.next_sibling.strip())
                R_flag=1
    
        if B_flag==0:
            Budget.append('NA')
        if R_flag==0:
            Revenue.append('NA')
        
        for x in soup.find_all('div',class_='summary_text'):
            Summary.append((x.get_text()).strip())

        P_flag=0
        for y in soup.find_all('div',class_='txt-block'):
            if  y.find('h4') and "Production Co:" in y.find('h4'):
                temp=y.find_all('span',itemprop='name')
                t1=list()
                for t in temp:
                    t1.append(t.get_text())
                Production.append(t1)
                P_flag=1
            
        if P_flag==0:
            Production.append('NA')
            
        D_flag=0
        W_flag=0
        S_flag=0

        for y in soup.find_all('div',class_='credit_summary_item'):
            if "Director:" in y.find('h4') or "Directors:" in y.find('h4'):
                temp = y.find('span',itemprop='name')
                Director.append(temp.get_text())
                D_flag=1
                
            if "Writer:" in y.find('h4') or "Writers:" in y.find('h4'):
                temp = y.find_all('span',itemprop='name')
                t1 = list()
                for t in temp:
                    t1.append(t.get_text())
                Writers.append(t1)
                W_flag=1
                
            if "Stars:" in y.find('h4') or "Star:" in y.find('h4'):
                temp = y.find_all('span',itemprop='name')
                t1 = list()
                for t in temp:
                    t1.append(t.get_text())
                Stars.append(t1)
                S_flag=1
                
        if D_flag==0:
            Director.append('NA')
        if W_flag==0:
            Writers.append('NA')
        if S_flag==0:
            Stars.append('NA')
```
